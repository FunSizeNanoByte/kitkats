##imputation using mean

```python
tips_data['total_bill'].isna()
```

```python
mean_total_bill=tips_data['total_bill'].mean()
#this takes all values (ignores NA) and gives you a mean value
mean_total_bill
```

```python
tips_data['total_bill']=tips_data['total_bill'].fillna(mean_total_bill)
tips_data.head(16)
```

```python
tips_data.fillna(mean_total_bill)
#fillna takes NA values and inputs what value you put (inside)
tips_data.head(16)
```
