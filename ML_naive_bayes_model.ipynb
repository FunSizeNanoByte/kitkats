```python
import pandas as pd
import sklearn.model_selection as ms
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split


```

```python
df=pd.read_csv("diabetes.csv")
df.head()
```

##contruct feature set and label

```python
feature=df.iloc[:,0:-1]
label=df['result']
feature.head()
```

```python
label.value_counts()
```

```python
# prompt: make 'feature' categorical variables into dummy variables. take out one column each. dont include 'age'

dm_feature=pd.get_dummies(feature[['gender','sudden_weight_loss',
                               'genital_thrush','visual blurring','itching','irritability'
                               ,'delayed_healing','alopecia']],drop_first=True)
dm_feature.head()
```

```python
dma_feature=pd.concat([feature[['age']],dm_feature],axis=1)
dma_feature.head()

```

```python
dm1_feature=pd.get_dummies(feature,columns=['obesity','gender','sudden_weight_loss',
                               'genital_thrush','visual blurring','itching','irritability'
                               ,'delayed_healing','alopecia'],drop_first=True)
dm1_feature.head()
#in example: its shorter! doesnt need comcat
```

construct training set and testing set

```python
train_x,test_x,train_y,test_y= train_test_split(dm1_feature,label,test_size=0.2,random_state=24)
#we are using 80% for training, 20% for testing against it. random state is a 'seed' number. for the same consistent random selections when needed
```

```python
len(train_x),len(test_x),len(train_y),len(test_y)
#train are the training models
#test are what well use to test the model.
```

construct navie bays model building

```python
NB_model=MultinomialNB()
NB_model.fit(train_x,train_y)
```

```python
pred_y=NB_model.predict(test_x)
pred_y
```

##model evaluation

```python
accuracy=accuracy_score(test_y,pred_y)
accuracy
```
