```python
# let's say we want to select a subset of the entire students and show distribution of
#math score, lunch, and further split into gender
subset=df.iloc[0:600,[0,3,5]]
```

```python
sns.swarmplot(data=subset,
              x='lunch',
              y='math_score',
              hue='gender',
              order=['standard',
              'free/reduced'],
              palette={'male':'green','female':'orange'})
plt.show()
```
