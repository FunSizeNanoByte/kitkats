```python
#this is a direct copy of the professors work, I claim no original content in this. it serves as teaching and example.
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('student.csv')
df.head()
```

```python
# extract all the categorical variables and put them into a
# new dataframe called cat_fts
cat_fts = df.iloc[:, 0:5]
cat_fts.head()
```

```python
# transform all the categorical variables to dummied variables
# put the dummied variables to a new dataframe called new_cats

new_cats = pd.get_dummies(cat_fts, columns=['gender', 'ethnicity',
                                            'parental_education', 'lunch',
                                            'test_preparation_course'],
                          drop_first=True)
new_cats.head()
```

```python
# select all numeric variables in df and put them into num_fts
num_fts = df.iloc[:, 5:8]
num_fts.head()
```

```python
# rescale all of the columns in num_fts using standarder
s = StandardScaler()
columns_to_normalize = ['math_score', 'reading_score', 'writing_score']
num_fts[columns_to_normalize] = s.fit_transform(num_fts[columns_to_normalize])
num_fts.head()
```

```python
# recreate an entire dataframe that contains all the values we transformed
# new_cats
# num_fts
# new_cats + num_fts
final_df = pd.concat([new_cats, num_fts], axis=1)
final_df.head()
```
