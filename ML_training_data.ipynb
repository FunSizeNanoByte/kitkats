```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
winedata = pd.read_csv("winequality.csv")
winedata.head()
```

##1 Please use the visualization tools in Python and show the relationship between "sulphates" and "fixed acidity". (Hint: You can use scatterplot for this question.)

```python
sns.jointplot(data=winedata,x='sulphates',y='fixed acidity',kind='reg')
#                                                                 /\Regression
plt.show()
```

##2. Use all the dataset as the training set, build the KNN model with K= 5.

```python
train_x=winedata.iloc[:,0:11]
train_y=winedata.iloc[:,-1]
train_x.head(5) #independent variables
```

```python
train_y.head(5) #dependant variable
```

```python
#setting models
model=KNeighborsClassifier(n_neighbors=5)
```

```python
model.fit(train_x,train_y)
```

##3. Suppose I got a bottle of Chateau Lafite in 1982. Its chemical substances are listed as follows:
lafite = [[4, 0, 0, 2, 0.1, 2, 6, 1, 3.5, 0.7, 8]]

```python
#prediction
lafite = [[4, 0, 0, 2, 0.1, 2, 6, 1, 3.5, 0.7, 8]]
model.predict(lafite)
#while I would love to get rid of this assigned feature warning,
#the result is 4!
```

##4. Please use your KNN model to classify which grade my Lafite belongs to?

The lafite has a predicted grade of 4!
